```python
'''39.  Write a python program to print the following pattern :
A B C D E
A B C D
A B C
A B
A'''

'''This means we want 5 rows in the pattern.
for i in range(n, 0, -1)

This is the outer loop.

It decides how many letters are printed per row.

range(n, 0, -1) â†’ starts at 5, stops before 0, moves by -1 each time.
So, values of i are

for j in range(i):

This is the inner loop.

It runs i times for each row.

Example:

When i = 5 â†’ j = 0, 1, 2, 3, 4 â†’ prints 5 letters

When i = 4 â†’ j = 0, 1, 2, 3 â†’ prints 4 letters

And so on.

print(chr(65 + j), end=" ")

Hereâ€™s the main logic ðŸ‘‡

chr() â†’ converts a number (ASCII value) to its character.

A has ASCII code 65, B = 66, C = 67, etc.'''


for i in range(n,0,-1): # start -n,stop-0,step -1
  for j in range(i):
    print(chr(65+j),end=" ")
  print()
```
